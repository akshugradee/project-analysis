# IPL DATA ANALYSIS 

```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl_data=pd.read_csv('iplmatches.csv')
```

```python
ipl_data.head()
```

```python
ipl_data['player_of_match'].value_counts()
```

```python
ipl_data['player_of_match'].value_counts()[0:10]
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl_data['player_of_match'].value_counts()[0:5].keys()),list(ipl_data['player_of_match'].value_counts()[0:5]),color='r')
plt.show()
```

```python
ipl_data['result'].value_counts()
```

```python
ipl_data['toss_winner'].value_counts()
```

```python
batting_1=ipl_data[ipl_data['win_by_runs']!=0]
```

```python
batting_1.head()
```

```python
plt.figure(figsize=(9,7))
plt.hist(batting_1['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
batting_1['winner'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(batting_1['winner'].value_counts()[0:3].keys()),list(batting_1['winner'].value_counts()[0:3]),color=['b','y','r'])
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.pie(list(batting_1['winner'].value_counts()),labels=list(batting_1['winner'].value_counts().keys()),autopct='%0.1f%%')
```
